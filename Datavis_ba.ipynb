```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
 import matplotlib.pyplot as plt
```

```python
bd=pd.read_csv('cars.csv',encoding='ISO-8859-1')
```

```python
# Assuming bd is your DataFrame and it has the columns 'mpg' (miles per gallon) and 'disp' (displaceme # Make sure to load your DataFrame first
# Create the joint plot 
sns.jointplot(x="mpg", y="disp", data=bd) 
plt.title('Joint Plot of MPG and Displacement') 
plt.show()
```

```python
sns.boxplot(data=bd, x='am', y='gear')  # replace with actual column names

# Optional: Set a title
plt.title('Cars', fontsize=16)

# Render and display the plot
plt.show()
```

```python
sns.barplot(x='gear', y='mpg', data=bd)

# Optional: Set a title
plt.title('cargear', fontsize=16)

# Render and display the plot
plt.show()
```
